Data Set
This dataset is connected to the research article, Changes in human peripheral blood mononuclear cell (HPBMC) populations and T-cell subsets associated with arsenic and polycyclic aromatic hydrocarbon exposures in a Bangladesh cohort. The researchers were Fredine T. Lauer, Faruque Parvez, Pam Factor-Litvak, Xinhua Liu, Regina M. Santella, Tariqul Islam, Mahbubul Eunus, Nur Alam, A. K. M. In this study, the hope was to add to the little data there is available regarding the effects of combined exposure to environmental arsenic (As) and polycyclic aromatic hydrocarbons (PAHs), specifically immunotoxicity to humans. To do this, 197 Bangladesh men were selected based on their exposure to As from drinking water and their cigarette smoking status (no women were studied because an insufficient amount were smokers). By design, 49% of the participants were smokers and 49% were exposed to water with arsenic levels over 50 ppb, or levels deemed unsafe in Bangladesh (the WHO recommended  limit is 10 ppb), creating the 197 rows/observations in the dataset. These men were surveyed and had blood and urine samples collected and the present immune cells were measured along with the T-helper cell subsets via an ex vivo activation method.This resulted in the 26 features that make up the dataset and are used throughout the paper: age (years), BMI, smoking status (yes or no), totally urinary adjusted for creatinine (Î¼g UAs/g creatinine), PAH exposure (per 108  nucleotides), the percent of T cells, T-helper cells, cytotoxic T lymphocytes, T memory cells, monocytes, non-monocyte (dose-dependent), B cells, activated B cells, natural killer cells, natural killer T cells, cells expressing the IL-7 receptor alpha, T regulatory cells, CD4 cells, Interferon cells, Interleukin-4 cells, interleukin-17 cells, stimulated CD4+ cells, stimulated CD3+ cells, and stimulated cells that are living.  

There were two other features used in the dataset that were essentially unmentioned in the report: water arsenic exposure and viability. These two features, while not directly applicable to the measurements used throughout the charts and report, are good explanations for any outliers. Any measurements that appear abnormal are likely explainable from one or both of these additional features.


Research Questions

The research question of this article was whether there are associations between As and PAH exposures with lymphoid cell populations in the peripheral blood mononuclear (PBMC) cells of humans. The objective was to examine alterations in the differentiation and activation of B and T cells. The hypothesis was that because large PAHs are carcinogenic and harmful to DNA, they will act synergistically with As in people. The features that were selected for this allow the assessment of association between As and PAH exposure, both separately and in combination, with damage to cell surface markers on PBMC and T-helper cell subsets.

Another question that could be assessed using this data is what changes in immune cell populations are unique to those exposed to harmful amounts of both As and PAH? Are there any immunotoxic consistencies that did not occur in anyone who measured safe levels in at least one of the carcinogens? Perhaps this could be a step in the direction of better understanding the coexistence of As and PAHs.

Figure 1 depicts the association between urinary arsenic concentrations and different cell surface markers. Each datapoint in figure one represents an observation and the solid lines represent the estimated mean outcome of living cells adjusted for other variables along with those in the model. The graphs act as a prime example of the apparent  inconsistency when not all the proper factors are considered. All four graphs have completely unique mean estimate lines due to how uniquely each marker is affected by As and PAH exposure.
